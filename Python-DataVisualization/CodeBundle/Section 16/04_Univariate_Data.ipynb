```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
sns.stripplot(y = 'tip', data=tips, jitter=False)
plt.ylabel('tips in USD')
plt.show()
```

```python
sns.stripplot(y = 'tip', data=tips, jitter=True)
plt.ylabel('tips in USD')
plt.show()
```

```python
sns.stripplot(x = 'day', y = 'tip', data=tips, jitter=True)
plt.ylabel('tips in USD')
plt.show()
```

```python
sns.stripplot(x = 'day', y = 'tip', data=tips, jitter=False)
plt.ylabel('tips in USD')
plt.show()
```

```python
sns.swarmplot(x = 'day', y = 'tip', data=tips)
plt.ylabel('tips in USD')
plt.show()
```

```python
sns.swarmplot(x = 'day', y = 'tip', data=tips, hue='sex')
plt.ylabel('tips in USD')
plt.show()
```

```python
sns.swarmplot(x = 'tip', y = 'day', data=tips, 
              hue='sex', orient='h')
plt.ylabel('tips in USD')
plt.show()
```

```python
sns.violinplot(x='day', y='tip', data=tips,
              inner=None, color='lightgray')
sns.stripplot(x='day', y='tip', data=tips,
              size=4, jitter=True)
plt.ylabel('tips in USD')
plt.show()
```
