```python
from bokeh.plotting import figure, output_file, show
from bokeh.layouts import gridplot
from bokeh.models import ColumnDataSource
from bokeh.models import CDSView, GroupFilter
```

```python
from bokeh.sampledata.iris import flowers
```

```python
source = ColumnDataSource(flowers)
```

```python
view1 = CDSView(source=source, 
                filters=[GroupFilter(column_name='species', 
                                     group='versicolor')])
```

```python
plot_size_and_tools = {'plot_height': 300, 'plot_width': 300,
                      'tools':['box_select', 'reset', 'help']}
```

```python
p1 = figure(title="Full data set", **plot_size_and_tools)
p1.circle(x='petal_length', y='petal_width', 
          source=source, color='black')
```

```python
show(p1)
```
