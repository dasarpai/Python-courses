```python
%matplotlib inline
import numpy as np
from bokeh.io import output_file, show
from bokeh.plotting import figure
from bokeh.layouts import column, row, gridplot
```

```python
x = list(range(11))
y0 = x
y1 = [10 - i for i in x]
y2 = [abs(i - 5) for i in x]
```

```python
s1 = figure(width=250, plot_height=250, 
            title=None)
s1.circle(x, y0, size=10, color='firebrick', 
          alpha=0.5)
```

```python
s2 = figure(width=250, height=250, 
            title=None)
s2.circle(x, y1, size=10, color='navy', 
          alpha=0.5)
```

```python
s3 = figure(width=250, height=250, 
            title=None)
s3.circle(x, y2, size=10, color='olive', 
          alpha=0.5)
```

```python
show(column(s1, s2, s3))
```

```python
show(row(s1, s2, s3))
```

```python
show(gridplot([[s1, s2],[None, s3]]))
```
