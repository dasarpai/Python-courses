```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 1000)
y = np.sin(x) 
np.random.random(1000) * 0.2
```

```python
plt.plot(x, y)
plt.show()
```

```python
from bokeh.io import output_file, show
from bokeh.plotting import figure
```

```python
plot = figure()
plot.line(x, y)
show(plot)
```

# Hover Appearance

```python
from bokeh.models import HoverTool
```

```python
#hover = HoverTool(tooltips=None, mode='hline')
hover = HoverTool(tooltips=None, mode='vline')
plot = figure(tools=[hover, 'crosshair'])
x = np.random.randn(100)
y = np.random.randn(100)
```

```python
# print(x, y)
plot.circle(x, y, size=15, hover_color='red')
```

```python
show(plot)
```

```python
from bokeh.models import Jitter
```

```python
plot = figure(plot_width=500, plot_height=400,
             x_range=(0, 3), y_range=(0, 10),
             title="Jitter Transform")
```

```python
y1 = np.random.random(2500) * 10
y2 = np.random.normal(size=2500) * 2 + 5
```

```python
plot.circle(x={'value': 1, 
               'transform': Jitter(width=0.4)},
              y=y1, color="navy", alpha=0.3)
```

```python
plot.circle(x={'value': 2, 
               'transform': Jitter(width=0.4)},
              y=y2, color="green", alpha=0.3)
```

```python
show(plot)
```
