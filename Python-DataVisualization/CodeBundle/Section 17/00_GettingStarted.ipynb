```python
%matplotlib inline
```

```python
from bokeh.io import output_file, show
```

```python
from bokeh.plotting import figure
```

```python
import numpy as np
```

```python
plot = figure(plot_width=400, 
              tools='pan, box_zoom')
```

```python
plot.circle(np.arange(5), [3, 4, 1, 9, 5])
```

```python
output_file('circle.html')
```

```python
show(plot)
```

```python
plot = figure()
plot.circle(x=[3, 5, 1, 9], 
            y=[1, 2, 3, 4], 
            size=[10, 20, 30, 40])
```

```python
show(plot)
```

```python
x = np.arange(5)
y = [2, 5, 9, 1, 8]
plot = figure()
plot.line(x, y, line_width=4)
show(plot)
```

```python
plot.line(x, y, line_width=4)
plot.circle(x, y, fill_color='red', 
            size=30)
show(plot)
```

```python
xs = [[1, 1, 2, 2], 
      [2, 2, 4], [2, 2, 3, 3]]
ys = [[2, 5, 5, 2], 
      [3, 5, 5], [2, 3, 4, 2]]
```

```python
plot = figure()
```

```python
plot.patches(xs, ys, 
fill_color=['yellow','magenta','cyan'],
line_color='black')
```

```python
show(plot)
```
