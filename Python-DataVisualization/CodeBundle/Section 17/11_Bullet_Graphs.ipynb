```python
from bokeh.io import show, output_notebook
from bokeh.palettes import PuBu4
from bokeh.plotting import figure
from bokeh.models import Label

output_notebook()
```

```python
data= [("Amit Jaiswal", 105, 120),
       ("Jignesh Shaha", 99, 110),
       ("Pankaj Jain", 109, 125),
       ("Pankaj Kothari", 135, 123),
       ("Saurav Maliwal", 45, 105)]

limits = [0, 20, 60, 100, 160]
labels = ["Poor", "OK", "Good", "Excellent"]
cats = [x[0] for x in data]
```

```python
p=figure(title="Sales Rep Performance", 
         plot_height=350, 
         plot_width=800, 
         y_range=cats)
p.x_range.range_padding = 0
p.grid.grid_line_color = None
p.xaxis[0].ticker.num_minor_ticks = 0
```

```python
zip(limits[:-1], limits[1:], PuBu4[::-1])
```

```python
for left, right, color in zip(limits[:-1], limits[1:], PuBu4[::-1]):
    p.hbar(y=cats, left=left, right=right, height=0.8, color=color)
show(p)
```

```python
perf = [x[1] for x in data]
p.hbar(y=cats, left=0, right=perf, 
       height=0.3, color="black")
show(p)
```

```python
comp = [x[2]for x in data]
p.segment(x0=comp, 
          y0=[(x, -0.5) for x in cats], 
          x1=comp,
          y1=[(x, 0.5) for x in cats], 
          color="white", 
          line_width=2)
show(p)
```

```python
for start, label in zip(limits[:-1], labels):
    p.add_layout(Label(x=start, y=0, text=label, 
                       text_font_size="10pt",
                       text_color='black', 
                       y_offset=5, x_offset=15))
show(p)
```
